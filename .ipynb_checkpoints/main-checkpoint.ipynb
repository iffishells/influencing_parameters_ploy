# Importing Libraries

```python
import pandas as pd
import numpy as np
import joblib
import pickle
```

# Input Data

```python
input_features = ["t", "Factor 1(P)", "Factor 2(T)", "Factor 4(W)", "Factor 5(R)"]
df = pd.read_excel("datasets/DataSet-testing.xlsx")
input_values = df[input_features]
```

# Trained Models

## Random Forest Model

```python

model_name = "random_forest_model"
with open(f"trained_models/{model_name}.pickle", "rb") as file:
    random_forest_model = pickle.load(file)
results_random_forest_model = pd.DataFrame( random_forest_model.predict(input_values) )
results_random_forest_model
```

## Gradient boosting Regressor Model

```python

model_name = "gradient_boosting_regressor"
with open(f"trained_models/{model_name}.pickle", "rb") as file:
    gradient_boosting_regressor = pickle.load(file)
results_gradient_boosting_regressor = pd.DataFrame( gradient_boosting_regressor.predict(input_values) )
results_gradient_boosting_regressor
```
